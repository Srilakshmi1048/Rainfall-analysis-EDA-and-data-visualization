```python
# Imputation method, replacing missing values with mean value of corresponding variable.

mydata['JAN'].fillna((mydata['JAN'].mean()), inplace=True)
mydata['FEB'].fillna((mydata['FEB'].mean()), inplace=True)
mydata['MAR'].fillna((mydata['MAR'].mean()), inplace=True)
mydata['APR'].fillna((mydata['APR'].mean()), inplace=True)
mydata['MAY'].fillna((mydata['MAY'].mean()), inplace=True)
mydata['JUN'].fillna((mydata['JUN'].mean()), inplace=True)
mydata['JUL'].fillna((mydata['JUL'].mean()), inplace=True)
mydata['AUG'].fillna((mydata['AUG'].mean()), inplace=True)
mydata['SEP'].fillna((mydata['SEP'].mean()), inplace=True)
mydata['OCT'].fillna((mydata['OCT'].mean()), inplace=True)
mydata['NOV'].fillna((mydata['NOV'].mean()), inplace=True)
mydata['DEC'].fillna((mydata['DEC'].mean()), inplace=True)
mydata['Jan-Feb'].fillna((mydata['Jan-Feb'].mean()), inplace=True)
mydata['Mar-May'].fillna((mydata['Mar-May'].mean()), inplace=True)
mydata['Jun-Sep'].fillna((mydata['Jun-Sep'].mean()), inplace=True)
mydata['Oct-Dec'].fillna((mydata['Oct-Dec'].mean()), inplace=True)
mydata['ANNUAL'].fillna((mydata['ANNUAL'].mean()), inplace=True)
```

# Trend Test(Moving average graph)

plotting a graph of yearly rainfall and add 10-year moving average line on that to check if there is a tendency of rainfall in india.

```python
fig = plt.figure(figsize = (14, 8))
tend =mydata.groupby('YEAR').mean()['ANNUAL']
tend.plot(kind='line', title = 'Overall Rainfall in Each Year', fontsize =20)
plt.ylabel("Overall Rainfall(mm)")
mydata['MA10'] = mydata.groupby('YEAR').mean()['ANNUAL'].rolling(10).mean()
mydata.MA10.plot()
```

when a 10 year moving average is plotted, we can see that rainfall has been decreasing trend  in the recent years.

In this above plot, from the 1900s to the 1960s, the amount of rainfall has increased but after then has decreased untill 2015.

```python
#Trend test: Kendall-tau test
import scipy.stats as stats
rainfall_rank = mydata.groupby('YEAR').mean()['ANNUAL']
rainfall_rank = pd.DataFrame({'Year':rainfall_rank.index, 'list':rainfall_rank.values})
rainfall_rank['rank'] = rainfall_rank['list'].rank(ascending = False)
t = list(range(1, len(rainfall_rank['list'])+1))
tau, p_value = stats.kendalltau(t, list(rainfall_rank['rank']))
print("tau: " + str(tau))
print("p-value: " + str(p_value))
```

```python
mydata[['YEAR', 'JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL',
       'AUG', 'SEP', 'OCT', 'NOV', 'DEC']].groupby("YEAR").sum().plot(figsize=(11,7));
```

From the result of Kendall-tau test, we can know there is no tendency in annual rainfall in india(p-value > 0.05, the alternative hypothesis will be rejected.)
